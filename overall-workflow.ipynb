## 1. Getting the data ready

```python
import pandas as pd

data_file = 'data/heart-disease.csv'
df = pd.read_csv(data_file)

df.head()
```

```python
df.describe()
```

```python
# Create X, features matix
X = df.drop("target", axis=1)
X.columns
```

```python
# Create y label
y = df['target']
```

## 2. Choose the right model and hyper parameters

```python
# Choosing Random Forest

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

## 3. Fit the model to the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
# make a prediction
y_pred = clf.predict(X_test)
```

## 4. Evaluate the model

```python
clf.score(X_test, y_test)
```

```python
# additional evaluations

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Classification Report")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

print("Accuracy:", accuracy_score(y_test, y_pred))
```

## 5. Improve a model

```python
# Try using diffrent hyper parameters for example: n_estimatros
import numpy as np

for i in range(10, 100, 5):
    print(f"Trying model with {i} estimators....")
    clf = RandomForestClassifier(n_estimators=i)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print(f"\tGot a score: {score * 100: 0.5f}%")
```

```python
clf = RandomForestClassifier(n_estimators=65)
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
score
```

## 6. Save this model and load it

```python
import pickle

pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
model = pickle.load(open('./random_forest_model_1.pkl', 'rb'))
model
```

```python
model.score(X_test, y_test)
```
